## Analyzing NYC Property Sales from 2003-2022

As any New Yorker can tell you, real estate in New York City is very expensive. According to [CBRE](https://www.cbre.com/singapore/about/media-centre/singapore-remains-the-2nd-most-expensive-housing-market-in-the-world-after-hong-kong), New York City is the 7th most expensive city in the world to buy a home. As a native New Yorker whose family is growing, space is becoming an issue and now browsing sites looking for the dream home is a daily occurance. While seeing today's prices for homes, I was curious how prices may have changed in the recent past given the great recession of 2008/09 and whether or not the old refrain about property values in NYC never going down is true.

The analysis seeks to answer questions on:
    
1. Have home values recovered from great recession?
2. Should I get a home, condo or coop?
3. Where would I get most bang for my buck? Comparing
